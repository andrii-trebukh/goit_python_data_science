```python
import string
import spacy
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize


nlp = spacy.load('en_core_web_sm')
```

```python
# Текст для обробки
with open("Hw12_data.txt", mode="r", encoding="utf8") as fd:
    text = fd.read()

# Токенізація
doc = nlp(text)
tokens = [token.text for token in doc]
print(tokens)
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')

tokens = word_tokenize(text)
sentences = sent_tokenize(text)
stop_words = set(stopwords.words('english'))
punctuation = string.punctuation + '\n'
```

```python
word_frequencies = {}
for word in doc:
  if word.text.lower() not in stop_words:
    if word.text.lower() not in punctuation:
      if word.text.lower() not in word_frequencies.keys():
        word_frequencies[word.text.lower()] = 1
      else:
        word_frequencies[word.text.lower()] += 1

sorted(word_frequencies, key=word_frequencies.get, reverse=True)
```

```python
# Нормалізуємо частоту слів:

max_frequency = max(word_frequencies.values())
for key, val in word_frequencies.items():
    word_frequencies[key] = val/max_frequency

word_frequencies
```

```python
# Підрахуємо оцінку кожного речення:

sentence_scores = {}

for sentence in sentences:
    for word in word_tokenize(sentence):
        if word.lower() in word_frequencies.keys():
            if sentence not in sentence_scores.keys():
                sentence_scores[sentence] = word_frequencies[word.lower()]
            else:
                sentence_scores[sentence] += word_frequencies[word.lower()]

sentence_scores
```

```python
# Виведемо перелік найкращих речень у порядку спадання оцінок.
from heapq import nlargest


select_length = int(len(sentences))
summary = nlargest(select_length, sentence_scores, key = sentence_scores.get)
summary
```
