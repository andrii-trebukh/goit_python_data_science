```python
import numpy as np
import pandas as pd
import seaborn as sns


data = pd.read_html(
    io="https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8",
    match="Коефіцієнт народжуваності в регіонах України",
    thousands=None,
    decimal=","
)

table = data[0]
table
```

```python
# Вивести перші рядки таблиці за допомогою методу head
table.head()
```

```python
# Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
table.shape
```

```python
# Замініть у таблиці значення "—" на значення NaN
table.replace("—", np.nan, inplace=True)
table
```

```python
# Визначте типи всіх стовпців за допомогою dataframe.dtypes
table.dtypes
```

```python
# Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
select = table.columns[table.isna().any()]
table[select] = table[select].astype("float")
table.dtypes
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
round((table.isnull().sum()/table.shape[0])*100, 2) # percentage of missing values in each column
```

```python
#Видаліть з таблиці дані по всій країні, останній рядок таблиці
table_ukraine = table[table["Регіон"] == "Україна"] # keep values for further use as diagram source data

index = table[table["Регіон"] == "Україна"].index
table.drop(index=index, axis=1, inplace=True)
table
```

```python
# Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
select = table.columns[table.isna().any()]
table[select] = table[select].fillna(table[select].mean())
table
```

```python
# Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
table[table["2019"] > table["2019"].mean()]["Регіон"].to_list()
```

```python
# У якому регіоні була найвища народжуваність у 2014 році?
table["Регіон"].iloc[table["2014"].idxmax()]
```

```python
# Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
sns.set_style("whitegrid")
ax = sns.barplot(
    x="Регіон",
    y="2019",
    data=table
)
ax.tick_params(axis='x', rotation=90)
ax.set_title("Коефіцієнт народжуваності по регіонах у 2019 році")
```

```python
# 2
# В мене закінчились ідеї як ще можна візуалізувати Коефіцієнт народжуваності, тому ось фігури Ліссажу

from matplotlib.pyplot import plot as plt
from math import pi

t=np.arange(0, 2 * pi, 0.001)

A = 1
B = 1
a1 = 3
b1 = 2
d = pi/2

a2 = 1
b2 = 2

X1 = A * np.sin(a1 * t + d)
Y1 = B * np.sin(b1 * t)

X2 = A * np.sin(a2 * t + d)
Y2 = B * np.sin(b2 * t)

plt(X1, Y1)
plt(X2, Y2)
```
