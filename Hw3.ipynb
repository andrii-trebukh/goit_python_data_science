```python
import pandas as pd
import numpy as np
```

```python
housing = pd.read_csv("Housing.csv")
housing.head()
```

```python
# напишіть функцію гіпотези лінійної регресії у векторному вигляді

def h(w, x):
    return np.dot(x, w)
```

```python
# створіть функцію для обчислення функції втрат у векторному вигляді

def mse(y_true, y_pred):
    return np.mean(np.sum((y_true - y_pred) ** 2))
```

```python
# реалізуйте один крок градієнтного спуску
# Gradient Descent

def grad_descent(x, y, iter=10000, lr=0.01, stopping_threshold=0.000001):
    w = np.zeros(x.shape[1])
    prev_cost = None
    
    for i in range(iter):
        y_pred = h(w, x)
        cost = mse(y, y_pred)

        if prev_cost and abs(cost - prev_cost) <= stopping_threshold:
            break
        
        prev_cost = cost

        derivative = (1 / x.shape[0]) * np.dot(x.T, (np.dot(x, w) - y))

        w -= lr * derivative
        print(f'Iteration {i+1}: Cost {cost}, Weights: {w}')
    
    return w
```

```python
# знайдіть найкращі параметри W для датасету прогнозуючу ціну на будинок залежно від площі, кількості ванних кімнат та кількості спалень



x = housing.iloc[:, 1:4]
x.insert(loc=0, column="ones", value=1)
x.area = x.area / 1000 # area in thousands units
print(x.head())

y = housing.price / 10 ** 6 # price in $ mln
print(y.head())

w = grad_descent(x, y)
print(f'Fianl weights: {w}')
```

```python
# знайдіть ці ж параметри за допомогою аналітичного рішення

w_analytical = np.dot(np.linalg.pinv(x), y)
w_analytical
```

```python
# порівняйте отримані результати

print(f"Gradient Descent: {w}")
print(f"Analytical solution: {w_analytical}")
print(f"Delta: {w - w_analytical}")
```
